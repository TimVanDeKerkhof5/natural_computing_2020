```python
#importing modules
import numpy as np 
import pandas as pd 

#loading datasets
data = pd.read_csv("../../../../../../../Documents/TMP/Natural Computing/breast-cancer-wisconsin-data/data.csv")
```

```python
#the fist 5 rows
data.head(5)
```

```python
#the dataset summary
data.info()
```

```python
#deleting useless columns
#deleting the "id" column
data.drop("id",axis=1,inplace=True)
#deleting the "Unnamed: 32" column
data.drop("Unnamed: 32",axis=1,inplace=True) 
```

```python
#the result
data.info()
```

```python
#the first 5 rows
data.head(5)
```

```python
#counting the diagnosis variable
data.diagnosis.value_counts()
```

```python
#diagnosis variable is a responsible variable for the classification
#replacing M and B with 1 and 0 respectively
data.diagnosis=data.diagnosis.map({'M':1,'B':0})
```

```python
#counting the diagnosis variable
data.diagnosis.value_counts()
```

```python
#finished the dataset preprocessing
#splitting dataset into training one and testing one
from sklearn.model_selection import train_test_split
train, test = train_test_split(data, test_size = 0.3,random_state=1234)
```

```python
#finding out the results
print(train.shape)
print(test.shape)
```

```python
#making independent variables for training
train_X = train.iloc[:, 1:31]
#making responsible variables for training
train_y=train.diagnosis
#making independent variables for testing
test_X= test.iloc[:, 1:31]
#making responsible variables for testing
test_y =test.diagnosis
```

```python
#finding out the results
print(train_X.shape)
print(train_y.shape)
print(test_X.shape)
print(test_y.shape)
```

# SVM

```python

#importing module
from sklearn import svm
#making the instance
model = svm.SVC(C= 6, kernel = 'linear',random_state=1)
#learning
model.fit(train_X,train_y)
#Prediction
prediction=model.predict(test_X)
#importing the metrics module
from sklearn import metrics
#evaluation(Accuracy)
print("Accuracy:",metrics.accuracy_score(prediction,test_y))
#evaluation(Confusion Metrix)
print("Confusion Metrix:\n",metrics.confusion_matrix(prediction,test_y))
```

```python
EstimatorDF = pd.DataFrame(columns=['n_estimators', 'accuracy_score', 'F1 micro', 'F1 macro', 'precision macro', 'precision micro', 'recall macro','recall micro'])
```

```python
import gc
from sklearn.ensemble import BaggingClassifier
N_E = [1,10,50,100]
for estimators in N_E:
    clf = BaggingClassifier(base_estimator=svm.SVC(C= 6, kernel = 'linear'), n_estimators=estimators, random_state=estimators, bootstrap = True).fit(train_X, train_y)
    clf.fit(train_X, train_y)
    pred = clf.predict(test_X)
    acc = metrics.accuracy_score(test_y,pred)
    f1_micro = metrics.f1_score(test_y,pred, average='micro')
    f1_macro = metrics.f1_score(test_y,pred,  average='macro')
    prec_macro = metrics.precision_score(test_y,pred, average='macro')
    prec_micro = metrics.precision_score(test_y,pred, average='micro')
    rec_macro = metrics.recall_score(test_y,pred, average='macro')
    rec_micro = metrics.recall_score(test_y,pred, average='micro')
    new_row = {'n_estimators' : estimators, 'accuracy_score': acc, 'F1 micro' : f1_micro, 'F1 macro' : f1_macro, 'precision macro':prec_macro, 'precision micro':prec_micro,'recall macro':rec_macro,'recall micro':rec_micro}
    #append row to the dataframe
    EstimatorDF = EstimatorDF.append(new_row, ignore_index=True)
    gc.collect()
EstimatorDF["n_estimators"] = EstimatorDF["n_estimators"].astype('int32')
print(EstimatorDF)
```

```python
EstimatorDF.to_csv('Breast.csv', index=False)
```
